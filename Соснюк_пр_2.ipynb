```python
# СОСНЮК КАТЕРИНА ФІТ 4-6 ВАРІАНТ 22
#Зчитування таблиці
import pandas as pd 
import numpy as np 

url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96" 
data = pd.read_html(url, header=0)[2]
```

```python
#1 Виведіть перші 5 рядків таблиці. 
df_1 = data.head(5)
print(df_1)
```

```python
#2 Визначте кількість рядків і стовпчиків за допомогою атрибуту shape 
print (data.shape)
```

```python
#3 Визначте загальну кількість років та кількість місяців у даних.
years = len(data['Рік']) - 1 
months = years*12  
print("Years total: ", years)
print("Months total: ", months)
```

```python
#4 Замінити пропущені значення  «-» в таблиці на NaN 

data = data.replace("—", np.nan)
data
```

```python
#5 Визначити тип стовпців за допомогою dtypes. 
column_types = data.dtypes
print("Типи стовпців:")
print(column_types)
```

```python
#6 Видаліть передостанній і останній стовпчики ( Кількість кгхліба[20] на середнюзарплату,  
#Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею. 
data = data.rename(columns={
    "Рік": "Year",
    "В національній валюті": "Salary in UAH",
    "В доларах США": "Salary in USD",
    "Кількість кг хліба[20] на середню зарплату": "Bread per Salary",
    "Кількість БігМаків на середню зарплату": "BigMacs per Salary"
})
data = data.drop(columns=["Bread per Salary", "BigMacs per Salary"])
data

```

```python
#7 Замінити тип нечислових даних на числові.
#ВИВОДИМО ТИПИ ДАНИХ ДО ПЕРЕТВОРЕННЯ
print("До перетворення")
data.dtypes
```

```python
#7 ПЕРЕТВОРЕННЯ
data["Year"] = data["Year"].str.replace("*", "")
data["Year"] = data["Year"].astype(int)

data["Salary in UAH"] = data["Salary in UAH"].str.replace('▲', "").str.replace('▼', "")
data["Salary in USD"] = data["Salary in USD"].str.replace('▲', "").str.replace('▼', "")
print("After")
data['Salary in UAH'] = data['Salary in UAH'].str.replace(',', '.').astype(float)
data['Salary in USD'] = data['Salary in USD'].str.replace(',', '.').astype(float)

```

```python
#7 ВИВОДИМО ТИПИ ДАНИХ ПІСЛЯ ПЕРЕТВОРЕННЯ
print("Після")
data.dtypes
```

```python
# 8 Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len) ПРОПУСКІВ НЕМАЄ
gaps = data.isnull().sum() / len(data)
print(gaps)
```

```python
# 9 Замініть всі пропущені значення на середні значення по (стовпцю). ПРОПУСКІВ НЕМАЄ АЛЕ ЗА ЇХ НА ЯВНОСТІ ТАКИЙ КОД ВИКОНАВ БИ ПОСТАВЛЕНЕ ЗАВДАННЯ
column_means = data.mean()

data.fillna(column_means, inplace=True)

print(data)
```

```python
#10 Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату. 
max_salary = data['Salary in UAH'].max()
print("Найбільша зарплата в грн: ", max_salary)
```

```python
#11 Знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату. 
min_salary = data['Salary in USD'].min()
min_index = data['Salary in USD'][data['Salary in USD'] == min_salary].index[0]
year = data.loc[min_index, 'Year']
print("Рік найменшої зарплати в доларах США: ", year)
print("Найменша зарплата в доларах США: ", min_salary)
```

```python
#12 Виведіть роки, коли зарплата в грн була вище середньої. 
average_salary = data['Salary in UAH'].mean()
years_above_average = data[data['Salary in UAH'] > average_salary]
print("Years with salaries higher than the average:")
print(years_above_average['Year'].tolist())
```

```python
#13 Порівняйте  зарплату  в грн за останні два доступних роки.
second_to_last_year = data.iloc[1]['Year']
salaries_last_two_years = data[data['Year'] >= second_to_last_year]
if salaries_last_two_years.iloc[0]['Salary in UAH'] > salaries_last_two_years.iloc[1]['Salary in UAH']:
    print("Зарплата в останньому році більша")
elif salaries_last_two_years.iloc[0]['Salary in UAH'] < salaries_last_two_years.iloc[1]['Salary in UAH']:
    print("Зарплата в останньому році менша")
else:
    print("Зарплати за останні два роки однакові")
```

```python
#14 Створіть графік, що відображає  зарплату в Україні з часом (за роками). 
import matplotlib.pyplot as plt
years = data['Year']
salaries = data['Salary in UAH']

plt.figure(figsize=(10, 6))
plt.plot(years, salaries, marker='o', linestyle='-')
plt.title('Зміна зарплати у гривнях з часом')
plt.xlabel('Рік')
plt.ylabel('Зарплата у гривнях')
plt.grid(True)
```
