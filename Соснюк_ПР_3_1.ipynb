```python
#5. Замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float.
column_means = data["video_views_for_the_last_30_days"].mean()
data["video_views_for_the_last_30_days"] = data["video_views_for_the_last_30_days"].fillna(column_means)
data["video_views_for_the_last_30_days"] = data["video_views_for_the_last_30_days"].astype(float)
column_means = data["country_rank"].mean()
data["country_rank"] = data["country_rank"].fillna(column_means)
data["country_rank"] = data["country_rank"].astype(float)
column_means = data["channel_type_rank"].mean()
data["channel_type_rank"] = data["channel_type_rank"].fillna(column_means)
data["channel_type_rank"] = data["channel_type_rank"].astype(float)
column_means = data["video_views_for_the_last_30_days"].mean()
data["video_views_for_the_last_30_days"] = data["video_views_for_the_last_30_days"].fillna(column_means)
data["video_views_for_the_last_30_days"] = data["video_views_for_the_last_30_days"].astype(float)
column_means = data["subscribers_for_last_30_days"].mean()
data["subscribers_for_last_30_days"] = data["subscribers_for_last_30_days"].fillna(column_means)
data["subscribers_for_last_30_days"] = data["subscribers_for_last_30_days"].astype(float)
column_means = data["Gross tertiary education enrollment (%)"].mean()
data["Gross tertiary education enrollment (%)"] = data["Gross tertiary education enrollment (%)"].fillna(column_means)
data["Gross tertiary education enrollment (%)"] = data["Gross tertiary education enrollment (%)"].astype(float)
column_means = data["Population"].mean()
data["Population"] = data["Population"].fillna(column_means)
data["Population"] = data["Population"].astype(float)
column_means = data["Unemployment rate"].mean()
data["Unemployment rate"] = data["Unemployment rate"].fillna(column_means)
data["Unemployment rate"] = data["Unemployment rate"].astype(float)
column_means = data["Urban_population"].mean()
data["Urban_population"] = data["Urban_population"].fillna(column_means)
data["Urban_population"] = data["Urban_population"].astype(float)
column_means = data["Latitude"].mean()
data["Latitude"] = data["Latitude"].fillna(column_means)
data["Latitude"] = data["Latitude"].astype(float)
column_means = data["Longitude"].mean()
data["Longitude"] = data["Longitude"].fillna(column_means)
data["Longitude"] = data["Longitude"].astype(float)
column_means = data["video views"].mean()
data["video views"] = data["video views"].fillna(column_means)
data["video views"] = data["video views"].astype(float)
column_means = data["uploads"].mean()
data["uploads"] = data["uploads"].fillna(column_means)
data["uploads"] = data["uploads"].astype(float)
column_means = data["highest_yearly_earnings"].mean()
data["highest_yearly_earnings"] = data["highest_yearly_earnings"].fillna(column_means)
data["highest_yearly_earnings"] = data["highest_yearly_earnings"].astype(float)
column_means = data["lowest_yearly_earnings"].mean()
data["lowest_yearly_earnings"] = data["lowest_yearly_earnings"].fillna(column_means)
data["lowest_yearly_earnings"] = data["lowest_yearly_earnings"].astype(float)
column_means = data["uploads"].mean()
data["highest_monthly_earnings"] = data["uploads"].fillna(column_means)
data["highest_monthly_earnings"] = data["highest_monthly_earnings"].astype(float)
column_means = data["lowest_monthly_earnings"].mean()
data["lowest_monthly_earnings"] = data["lowest_monthly_earnings"].fillna(column_means)
data["lowest_monthly_earnings"] = data["lowest_monthly_earnings"].astype(float)
data
```

```python
#8. Побудуйте діаграму розподілу переглядів (використовуйте kind='hist'):
import matplotlib.pyplot as plt
data['video views'].plot(kind='hist', bins=5, color='blue', edgecolor='black', alpha=0.7)

plt.title('Діаграма розподілу переглядів')
plt.xlabel('Прегляди')
plt.ylabel('Частота')
plt.grid(True)

plt.show()
```

```python
#9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean) 
maximum_views = data['video views'].max()
minimum_views = data['video views'].min()
average_views = data['video views'].mean()

print("Максимальна кількість переглядів:", maximum_views)
print("Мінімальна кількість переглядів:", minimum_views)
print("Середня кількість переглядів:", average_views)
```

```python
#10.Знайдіть країну, де найбільше відео було завантажено на YouTube
country_stats = data.groupby('Country')['uploads'].sum().reset_index()
most_uploaded_country = country_stats.loc[country_stats['uploads'].idxmax()]
print("Країна, де найбільше відео було завантажено на YouTube:", most_uploaded_country['Country'])
```

```python
#11. Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads. 
highest_upload_idx = data['uploads'].idxmax()
highest_upload_video = data.loc[highest_upload_idx, 'Title']

lowest_upload_idx = data['uploads'].idxmin()
lowest_upload_video = data.loc[lowest_upload_idx, 'Title']

print("Відео з найбільшою кількістю завантажень:", highest_upload_video)
print("Відео з найменшою кількістю завантажень:", lowest_upload_video)
```
