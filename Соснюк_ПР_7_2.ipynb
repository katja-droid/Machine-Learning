```python
# СОСНЮК КАТЕРИНА ФІТ 4-6 ВАРІАНТ 22 
# Загрузити датасет, що містить інформацію про відвідуваність музеїв в різні роки. 
# Вивести перші 5 рядків датасету. 
import pandas as pd
import csv
xls_file = pd.ExcelFile('museum_visitors.xlsx')
df = xls_file.parse(xls_file.sheet_names[0])
df.to_csv('output.csv', index=False)
with open('output.csv', 'r') as file:
    csv_reader = csv.reader(file)
    data = list(csv_reader)
column_names = []
for row in data[0]:
    column_names.append(row.split(','))
column_names
data_rows = []
for row in data:
    if data.index(row)!=0:
        data_rows.append(row[0].split(','))
df = pd.DataFrame(data_rows, columns=column_names[0])
df.head(5)
```

```python
# Провести попередній аналіз даних (визначити розмір датасету, тип даних, кількість пропусків). 
df.shape
```

```python
missing_values = df.isna().sum()
missing_values
#ПРОПУСКІВ НЕМАЄ
```

```python
df.dtypes
#ТИПИ ДАНИХ object
```

```python
df['Avila Adobe'] = df['Avila Adobe'].astype(float)
df['Firehouse Museum'] = df['Firehouse Museum'].astype(float)
df['Chinese American Museum'] = df['Chinese American Museum'].astype(float)
df['America Tropical Interpretive Center'] = df['America Tropical Interpretive Center'].astype(float)
df
```

```python
# Перевірити наявність дублікатів. Якщо є, видалити їх і вивести розмір таблиці (без дублікатів). 
duplicates = df[df.duplicated()]
if not duplicates.empty:
    df = df.drop_duplicates()
else:
    print("Дублікатів немає.")
```

```python
# Про які роки містить інформацію датасет.
df['Date'] = pd.to_datetime(df['Date'])
unique_years = df['Date'].dt.year.unique()
unique_years
```

```python
# Замінити назви столбців. Замінити великі букви на маленьки, пробіли на підкреслення 
# (date, avila_adobe, firehouse_museum, chinese_american_museum, america_tropical_interpretive_center)
df.rename(columns={'Date': 'date', 'Avila Adobe': 'avila_adobe', 'Firehouse Museum':'firehouse_museum','Chinese American Museum':'chinese_american_museum', 'America Tropical Interpretive Center': 'america_tropical_interpretive_center'}, inplace=True)
df.head()
```

```python
# Обчислити середню кількість відвідувачів для кожного музею протягом всього періоду.
df_copy = df.copy()
df_copy['date'] = pd.to_datetime(df_copy['date'])
df_copy.set_index('date', inplace=True)
average_visitors = df_copy.mean()
print(average_visitors)
```

```python
# Знайти мінімальну та максимальну кількість відвідувачів для кожного музею за 2018 рік 
df_2018 = df[df['date'].dt.year == 2018]
min_visitors_2018 = df_2018.min()
max_visitors_2018 = df_2018.max()

print("Мінімальна к-сть відвідувачів у 2018:")
print(min_visitors_2018)

print("\Максимальна к-сть відвідувачів у 2018:")
print(max_visitors_2018)
```

```python
# Визначити місяці з найвищою і найнижчою загальною кількістю відвідувачів серед усіх музеїв для 2015 року. Вивести назви місяців (не цифри). 
df_2015 = df[(df['date'].dt.year == 2015)]

monthly_visitors = df_2015.groupby(df_2015['date'].dt.strftime('%B'))['avila_adobe', 'firehouse_museum', 'chinese_american_museum','america_tropical_interpretive_center'].sum()
highest_month = monthly_visitors.sum(axis=1).idxmax()
lowest_month = monthly_visitors.sum(axis=1).idxmin()

print(f"Найвища загальна к-сть відвідувачів у 2015 (місяць): {highest_month}")
print(f"Найнижча загальна к-сть відвідувачів у 2015 (місяць): {lowest_month}")
```

```python
# Порівняти кількість відвідувачів музею "Avila Adobe" у літні і зимові місяці 2018 року
summer_months = ['06', '07', '08']
winter_months = ['12']

summer_data = df[df['date'].dt.strftime('%m').isin(summer_months)]
winter_data = df[df['date'].dt.strftime('%m').isin(winter_months)]

total_visitors_summer = summer_data['avila_adobe'].sum()
total_visitors_winter = winter_data['avila_adobe'].sum()
print(f"Кількість відвідувачів у літні місяці 2018: {total_visitors_summer}")
print(f"Кількість відвідувачів у зимові місяці 2018: {total_visitors_winter}")
#У ЛІТНІ МІСЯЦІ ВІДВІДУВАНІСТЬ ЗНАЧНО БІЛЬША
```

```python
# Знайти кореляцію між кількістю відвідувачів в кожному з музеїв  та датами у 2016 році 
df_copy = df.copy()
df_copy['date'] = df_copy['date'].apply(lambda x: x.toordinal())
correlation_matrix = df_copy.corr()
date_correlation = correlation_matrix['date'][1:]
print("Кореляція між кількістю відвідувачів в кожному з музеїв  та датами у 2016 році:")
print(date_correlation)
```

```python
# Побудувати графік ліній для кожного музею, показуючи їхню відвідуваність протягом 2017 року
import matplotlib.pyplot as plt
df_2017 = df[(df['date'].dt.year == 2017)]
df_2017.set_index('date', inplace=True)
plt.figure(figsize=(10, 6))  

for column in df_2017.columns:
    plt.plot(df_2017.index, df_2017[column], label=column)

plt.title('Відвідуваність музеїв протягом 2017')
plt.xlabel('Дата')
plt.ylabel('Відвідуваність')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

```python
# Побудувати графіки розсіювання для кожного музею за 2018 рік 
df_2018 = df[(df['date'].dt.year == 2018)]
plt.figure(figsize=(10, 6))  

for column in df_2018.columns[1:]:
    plt.scatter(df_2018['date'], df_2018[column], label=column, s=10) 
plt.title('Відвідуваність музеїв протягом 2018')
plt.xlabel('Дата')
plt.ylabel('Відвідуваність')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

```python
# Гістограма відвідуваності за місяцями кожного музею. Вивести 4 графіки: за 2014, 2015, 2016, 2017 роки. 
# Графіки розташовані два зверху, два знизу, колір першого графіка - синій, другого зелений, третього жовтий, четвертого сірий. 
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
colors = ['blue', 'green', 'yellow', 'gray']

for i, year in enumerate([2014, 2015, 2016, 2017]):
    ax = axs[i // 2, i % 2] 
    year_data = df[df['date'].dt.year == year]  
    for column, color in zip(year_data.columns[1:], colors):
        ax.hist(year_data[column], bins=12, color=color, alpha=0.7, label=column)

    ax.set_title(f'Відвідуваність за місяцями у {year}')
    ax.set_xlabel('Кількість відвідувань')
    ax.set_ylabel('Частота')
    ax.legend()

plt.tight_layout()
plt.show()
```
